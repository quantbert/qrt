Hand-rolled feature primitives, one file per feature under `qrt/feat/qta/`.
These code cells are **executed live** by Quarto every time the docs are built.

## Sample data

We use qrt's bundled AAPL sample dataset — loaded offline via `q.data.datasets.load`,
no network dependency (see the [Data](../data/index.qmd) tutorial for more on
`q.data`):

```python
import pandas as pd

import qrt as q

aapl = q.data.datasets.load("aapl")
aapl.tail()
```

Hand-rolled primitives, one file per feature under `qrt/feat/qta/`:

```python
sma_20 = q.feat.qta.sma(aapl["close"], 20)
sma_100 = q.feat.qta.sma(aapl["close"], 100)

price_features = pd.DataFrame({"close": aapl["close"], "sma_20": sma_20, "sma_100": sma_100})
price_features.tail()
```

```python
q.feat.qta.lags(aapl["close"], [1, 5, 21]).tail()
```

To add a feature, create `qrt/feat/qta/<feature>.py` with one public function and
re-export it in `qrt/feat/qta/__init__.py` (one import line + one `__all__` entry).

## Visualizing features

`q.plot.col` renders any feature DataFrame as an interactive Plotly chart with
hover, zoom, and range-selector buttons — handy for eyeballing an indicator
against price before using it in a model:

```python
fig = q.plot.col(
    price_features,
    title="AAPL close price with SMA overlays",
    ylabel="Price",
    height=450,
)
fig.show()
```
