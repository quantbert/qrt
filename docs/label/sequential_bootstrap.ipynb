`q.label.sequential_bootstrap` draws events in proportion to the marginal uniqueness they would contribute if selected next. Probabilities are recomputed after every draw.

Like an ordinary bootstrap, events may repeat. Pass `random_state` for reproducible bags.

```python
import pandas as pd

import qrt as q

observations = pd.date_range("2026-01-01", periods=10, name="datetime")
end_times = pd.Series(
    observations[[5, 6, 7, 9]],
    index=observations[[0, 1, 4, 8]].rename("event_time"),
)
sample = q.label.sequential_bootstrap(
    observations,
    end_times,
    size=12,
    random_state=42,
)
pd.Series(sample, name="sampled_event").to_frame()
```

```python
pd.Series(sample).value_counts().rename("draws").sort_index().to_frame()
```
