`q.label.indicator_matrix` builds a sparse observation-by-event membership matrix. A value of 1 means the observation belongs to the event's inclusive lifetime.

Rows sum to label concurrency. Columns expose each event path for uniqueness calculations and sequential bootstrap sampling.

```python
import pandas as pd

import qrt as q

observations = pd.date_range("2026-01-01", periods=7, name="datetime")
end_times = pd.Series(
    observations[[3, 5, 6]],
    index=observations[[0, 1, 4]].rename("event_time"),
)
membership = q.label.indicator_matrix(observations, end_times)
membership.sparse.to_dense()
```

```python
pd.DataFrame({
    "matrix row sum": membership.sum(axis=1),
    "concurrency": q.label.concurrency(observations, end_times),
})
```
