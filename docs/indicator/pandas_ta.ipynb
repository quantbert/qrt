`q.indicator.pandas_ta` wraps every [pandas-ta-classic](https://xgboosted.github.io/pandas-ta-classic/) indicator. The provider path preserves pandas-ta's formulas, parameters, and output names explicitly.

## Sample data

We use qrt's bundled AAPL sample dataset — loaded offline via `q.data.datasets.load`,
no network dependency (see the [Data](../data/index.qmd) tutorial for more on
`q.data`):

```python
import pandas as pd

import qrt as q

aapl = q.data.datasets.load("aapl")
aapl.tail()
```

Outputs keep pandas-ta's parameterised names (e.g. `BBL_20_2.0`, `RSI_14`):

```python
bbands = q.indicator.pandas_ta.bbands(aapl, length=20)
bbands.tail()
```
