`q.indicator.returns` calculates the percentage change in one instrument's price over a fixed number of observations. It preserves the input index and leaves the first `periods` rows missing because no earlier price is available.

The result is a decimal return: `0.01` means a 1% gain.

```python
import pandas as pd

import qrt as q

data = q.data.datasets.load("aapl")
end = data.index.max()
data = data.loc[end - pd.DateOffset(years=1) :].copy()
daily_returns = q.indicator.returns(data["close"])
pd.DataFrame({"close": data["close"], "return": daily_returns}).tail()
```

## Inspect the return series

Plotting returns in percent makes the scale easier to read while leaving the indicator itself in decimal form.

```python
figure = q.plot.col(
    daily_returns.mul(100).to_frame("AAPL daily return"),
    title="AAPL close-to-close returns",
    ylabel="Return (%)",
)
figure.show(renderer="notebook_connected")
```
