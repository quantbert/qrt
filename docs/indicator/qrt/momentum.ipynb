`q.indicator.momentum` measures the percentage change from the close `window` observations ago. Positive values indicate appreciation over the lookback and negative values indicate depreciation.

Momentum is a measurement, not a trading signal. The initial window is missing.

```python
import pandas as pd

import qrt as q

prices = pd.concat(
    {"AAPL": q.data.datasets.load("aapl")["close"], "SPY": q.data.datasets.load("spy")["close"]},
    axis=1,
    join="inner",
).dropna()
end = prices.index.max()
prices = prices.loc[end - pd.DateOffset(years=3) :]
prices.tail()
```

## Compare instruments

A 63-session window is roughly one trading quarter. Calling the function column by column keeps instrument boundaries explicit.

```python
window = 63
quarterly_momentum = prices.apply(
    lambda series: q.indicator.momentum(series, window=window)
).mul(100)
quarterly_momentum.tail()
```

```python
figure = q.plot.col(
    quarterly_momentum.add_suffix(f" {window}-session momentum"),
    title="AAPL and SPY quarterly price momentum",
    ylabel="Momentum (%)",
)
figure.show(renderer="notebook_connected")
```
