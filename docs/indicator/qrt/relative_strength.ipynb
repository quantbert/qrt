`q.indicator.relative_strength` subtracts a benchmark's lookback return from an asset's lookback return. Positive values mean the asset outperformed the benchmark over that window; negative values mean it underperformed.

This example measures AAPL against SPY over 63-session lookbacks during the latest five shared years. The function aligns SPY to AAPL's index without filling missing dates.

```python
import pandas as pd

import qrt as q

prices = pd.concat(
    {
        "AAPL": q.data.datasets.load("aapl")["close"],
        "SPY": q.data.datasets.load("spy")["close"],
    },
    axis=1,
    join="inner",
).dropna()
end = prices.index.max()
prices = prices.loc[end - pd.DateOffset(years=5) :]
prices.tail()
```

## Calculate the indicator

Pass the asset and benchmark explicitly. `lookback=63` is approximately one trading quarter for daily observations.

```python
lookback = 63
strength = q.indicator.relative_strength(
    prices["AAPL"],
    prices["SPY"],
    lookback=lookback,
)
strength.tail()
```

## Compare with SPY

SPY is already part of the indicator calculation. The zero line separates AAPL outperformance from underperformance.

```python
comparison = pd.DataFrame(
    {
        "AAPL relative strength vs SPY": strength.mul(100),
        "Equal performance": 0.0,
    }
)
fig = q.plot.col(
    comparison,
    title=f"AAPL relative strength versus SPY ({lookback} sessions)",
    ylabel="Return difference (percentage points)",
)
fig.show(renderer="notebook_connected")
```
