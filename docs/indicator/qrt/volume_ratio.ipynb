`q.indicator.volume_ratio` divides each volume observation by a trailing mean or median baseline. A ratio of 2 means volume is twice its rolling baseline; values below 1 indicate quieter trading.

The median is the default because isolated volume spikes affect it less than the mean.

```python
import pandas as pd

import qrt as q

data = q.data.datasets.load("aapl")
end = data.index.max()
data = data.loc[end - pd.DateOffset(years=2) :].copy()
median_ratio = q.indicator.volume_ratio(data["volume"], window=20)
mean_ratio = q.indicator.volume_ratio(data["volume"], window=20, statistic="mean")
ratios = pd.DataFrame({"median baseline": median_ratio, "mean baseline": mean_ratio})
ratios.tail()
```

## Compare rolling baselines

Large observations raise the rolling mean, which can suppress subsequent mean-based ratios. The median-based series remains more responsive after an isolated spike.

```python
figure = q.plot.col(
    ratios,
    title="AAPL volume relative to its 20-session baseline",
    ylabel="Volume ratio",
)
figure.add_hline(y=1, line_dash="dot", line_color="#555555")
figure.show(renderer="notebook_connected")
```
