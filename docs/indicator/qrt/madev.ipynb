`q.indicator.madev` measures the trailing mean absolute distance between one instrument's close and its simple moving average. The reviewed formula uses the same window twice, so the first defined value requires $2n - 1$ observations.

The function accepts one ordered `Series`, preserves its index, and does not mutate the input.

```python
import pandas as pd

import qrt as q

data = q.data.datasets.load("aapl")
end = data.index.max()
data = data.loc[end - pd.DateOffset(years=5) :].copy()
window = 20
data["madev"] = q.indicator.madev(data["close"], window=window)
data[["close", "madev"]].tail()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

figure = make_subplots(specs=[[{"secondary_y": True}]])
figure.add_trace(
    go.Scatter(x=data.index, y=data["close"], name="AAPL close", line={"color": "#2457a7"}),
    secondary_y=False,
)
figure.add_trace(
    go.Scatter(x=data.index, y=data["madev"], name=f"MADEV ({window})", line={"color": "#d04a35"}),
    secondary_y=True,
)
figure.update_layout(title="AAPL close and deviation from its rolling SMA", hovermode="x unified")
figure.update_yaxes(title_text="Close", secondary_y=False)
figure.update_yaxes(title_text="MADEV", secondary_y=True)
figure.show(renderer="notebook_connected")
```
