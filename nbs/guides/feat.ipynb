# Feature engineering

`q.feat` collects feature engineering tools under one namespace, organized as submodules:

| Submodule | What it is |
|---|---|
| `q.feat.qta` | our own hand-rolled features, one file per feature |
| `q.feat.talib` | every [TA-Lib](https://ta-lib.github.io/ta-lib-python/) indicator, pandas-friendly |
| `q.feat.pandas_ta` | every [pandas-ta-classic](https://xgboosted.github.io/pandas-ta-classic/) indicator |

All indicator functions share the same convention: they accept an **OHLCV DataFrame**
with lowercase columns (`open, high, low, close, volume`) -- the frame layout returned
by `q.vendors` -- and preserve the datetime index on the output. Indicators that only
need a price series also accept a plain `Series` (treated as `close`).

## qta -- our own features

```python
q.feat.qta.sma(ohlc["close"], 20)
q.feat.qta.lags(ohlc["close"], [1, 5, 21])
```

To add a feature, create `nbs/feat/qta/<feature>.ipynb` with `#| default_exp feat.qta._<feature>`
and one public function, then re-export it in `nbs/feat/qta/02_init.ipynb`
(one import line + one `__all__` entry).

## TA-Lib

```python
q.feat.talib.RSI(ohlc)                    # Series named 'rsi'
q.feat.talib.ATR(ohlc, timeperiod=20)     # uses high/low/close
q.feat.talib.MACD(ohlc)                   # DataFrame: macd, macdsignal, macdhist
q.feat.talib.SMA(prices, timeperiod=20)   # Series in, Series out
```

Single-output indicators return a `Series` named after the indicator; multi-output
indicators return a `DataFrame` with TA-Lib's output names as columns.

## pandas-ta-classic

```python
q.feat.pandas_ta.rsi(ohlc)                  # Series 'RSI_14'
q.feat.pandas_ta.bbands(ohlc, length=20)    # DataFrame BBL/BBM/BBU...
q.feat.pandas_ta.cdl_pattern(ohlc, name="doji")
```

Outputs keep pandas-ta's parameterised names (e.g. `RSI_14`, `MACD_12_26_9`).

## IDE support

The `talib` and `pandas_ta` wrappers create their functions dynamically at runtime
(via module-level `__getattr__`), which static analyzers (Pylance, mypy) can't see.
Autocomplete and hover docs are provided by hand-maintained type stubs -- regenerate
them after upgrading either library:

```bash
make stubs
```

The full, exhaustive indicator list is also browsable in the
[API reference](../api/qrt/feat.html), which imports the live modules and can see
the dynamically-created functions.
